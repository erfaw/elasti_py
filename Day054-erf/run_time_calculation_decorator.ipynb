```python
# Build a speed function cacluator decorator
import time
def speed_calc_decorator(f):
    def wrap_f():
        start_run = time.time()
        f()
        end_run = time.time()
        whole_run_time = end_run - start_run
        print(f"\nrun time: <{whole_run_time}>s")
    return wrap_f
```

```python
# A simple function to test decorator
@speed_calc_decorator
def print_100_hello():
    for _ in range(5):
        time.sleep(1)
        print("hello", end=", ")
```

```python
# Run to see result
print_100_hello()
```
