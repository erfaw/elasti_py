```python
# Make decorator
import time
def delay_decorator(function):
    def wrapped_f():
        print("w8 for 2s...")
        time.sleep(2)
        function()
    return wrapped_f
```

```python
# Base function without sugar decorator syntax
def say_hello():
    print("Hello there!")
```

```python
# Replace base function with decorador MANUALly
say_hello = delay_decorator(say_hello)
```

```python
# Base function with sugar decorator syntax
@delay_decorator
def say_hello():
    print("Hello there!")
```

```python
# Run function after decorating
say_hello()
```
