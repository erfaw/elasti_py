```python
# Create a logging_decorator() which is going to print the name of the function that was called, the arguments it was given and finally the returned output 
```

```python
def logging_deorator(f):
    def wrapper(*args):
        print(f"function name => {f.__name__}\nfunction args => {args}\nfunction output => {f(*args)}")
        return 
    return wrapper
```

```python
@logging_deorator
def a_function(*args):
    sum = 0
    for arg in args:
        sum += arg
    return sum
```

```python
a_function(1,2,3)
```
