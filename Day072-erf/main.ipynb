```python
# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?
df[
    df['Mid-Career Median Salary'] == df['Mid-Career Median Salary'].min()
]
```

```python
spread= df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
```

```python
df.insert(len(df.columns), "spread", spread, )
```

```python
# if the difference between the 10th percentile and the 90th percentile earnings of your major is small, then you can be more certain about your salary after you graduate.
df[
    df["spread"] == df["spread"].min()
]
```

```python
df.sort_values("spread", ascending= True)
```

```python
df.sort_values("spread", ascending= True)[['Undergraduate Major', 'spread']].head()
```

```python
# Find the top 5 degrees with the highest values in the 90th percentile.
df.sort_values(
    'Mid-Career 90th Percentile Salary',
    ascending= False
).head()
```

```python
# find the degrees with the greatest spread in salaries.
spread_starting_and_mid = df['Mid-Career Median Salary'] - df['Starting Median Salary']
df.insert(len(df.columns), "diff mid and starter", spread_starting_and_mid)
```

```python
df.sort_values(
    'diff mid and starter',
    ascending= False
).head()
```
